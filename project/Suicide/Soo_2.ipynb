```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
physical_activity =  "../../data/Physical_activity.csv"
physical_activity_df = pd.read_csv(physical_activity)
physical_activity_df.head()
```

```python
physical_activity_clean = physical_activity_df[["Year","LocationAbbr", "Question", "Data_Value", "Gender"]]
physical_activity_clean = physical_activity_clean.loc[physical_activity_clean['Year']==2020]
physical_activity_clean = physical_activity_clean.groupby("Gender")
physical_activity_clean.head()
```

```python
states = ["ME","SD","NJ","GA","MS"]
mean_value = []

# physical_by_State
for state in states:
     state_match_s = physical_activity_clean["LocationAbbr"] == state
     state_match_value_s = physical_activity_clean[state_match_s,'LocationAbbr']    
    # add subset to tumor_vol_data
mean_value.append(state_match_value_s)

mean_value
```
