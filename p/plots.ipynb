```python
suicide =  "../data/Suicide.csv"
suicide_df = pd.read_csv(suicide)
suicide_df = suicide_df.loc[suicide_df['YEAR']==2020]
suicide_df.head()
```

```python
len(suicide_df)
```

```python

# Filter the DataFrame down only to those columns to chart
suicide_clean = suicide_df[["STATE","RATE"]]

# Set the index to be "State" so they will be used as labels
suicide_clean = suicide_clean.set_index("STATE")

suicide_clean.head()
```

```python
# Use DataFrame.plot() in order to create a bar chart of the data
suicide_clean.plot(kind="bar", figsize=(20,5))

# Set a title for the chart
plt.title("Suicide Rate per State")

plt.tight_layout()
plt.show()
```
